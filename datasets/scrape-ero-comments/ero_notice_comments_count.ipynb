```python
import pandas as pd
import json
import matplotlib.pyplot as plt

notice_comments_count_json_file = open('output/ero-notices-comments-counts.json')
notice_comments_json = json.load(notice_comments_count_json_file)
df = pd.DataFrame.from_dict(notice_comments_json, orient='index')
sorted_by_registry_df = df.sort_values(by=['comments_from_registry'], ascending=False)[:20]
sorted_by_email_df = df.sort_values(by=['comments_by_email'], ascending=False)[:20]
sorted_by_mail_df = df.sort_values(by=['comments_by_mail'], ascending=False)[:20]
```

```python
print(sorted_by_registry_df['comments_from_registry'])
```

```python
print(sorted_by_email_df['comments_by_email'])
```

```python
print(sorted_by_mail_df['comments_by_mail'])
```

```python
sorted_by_registry = sorted_by_registry_df['comments_from_registry'].plot.barh(rot=0, title='Top 20 notices with comments submitted through ERO registry')
sorted_by_registry.invert_yaxis()
sorted_by_registry.set_xlabel("Comment count")
sorted_by_registry.set_ylabel("ERO Notice")
```

```python
sorted_by_email = sorted_by_email_df['comments_by_email'].plot.barh(rot=0, title='Top 20 notices with comments submitted by email')
sorted_by_email.invert_yaxis()
sorted_by_email.set_xlabel("Comment count")
sorted_by_email.set_ylabel("ERO Notice")
```

```python
sorted_by_mail = sorted_by_mail_df['comments_by_mail'].plot.barh(rot=0, title='Top 20 notices with comments submitted by mail')
sorted_by_mail.invert_yaxis()
sorted_by_mail.set_xlabel("Comment count")
sorted_by_mail.set_ylabel("ERO Notice")
```
